```
#import libraries
import pandas as pd 
from pandas_datareader import data
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
plt.style.use('fivethirtyeight')
```

```
#load the data from Yahoo finance
start_date = '2015-01-09'
end_date = '2019-01-01'
ibm = data.DataReader('IBM', 'yahoo', start_date, end_date)
ibm.tail(10)
```

```
#visualize the data
plt.figure(figsize=(12,5))
plt.plot(ibm['Adj Close'],label='IBM')
plt.xlabel('first of 2015 till the first of 2019')
plt.title('IBM adj Close price History')
plt.ylabel('Adj Close Price in $')
plt.legend(loc='upper right')
plt.show()
```

```
#Create a simple moving average with a 50 days window
SMA50=pd.DataFrame()
SMA50['Adj Close']=ibm['Adj Close'].rolling(window=50).mean()
SMA50
```

```
#Create a simple moving average of 100 days window
SMA100=pd.DataFrame()
SMA100['Adj Close']=ibm['Adj Close'].rolling(window=100).mean()
SMA100
```

```
#Visualize the data
plt.figure(figsize=(12,7))
plt.plot(ibm['Adj Close'],label='IBM')
plt.plot(SMA50['Adj Close'],label='SMA50')
plt.plot(SMA100['Adj Close'],label='SMA100')
plt.xlabel('first of 2015 till the first of 2019')
plt.title('IBM adj Close price History')
plt.ylabel('Adj Close Price in $')
plt.legend(loc='upper right')
plt.show()
```

```
#Create a new Dta frame to store all the data
new_data=pd.DataFrame()
new_data['IBM']=ibm['Adj Close']
new_data['SMA50']=SMA50['Adj Close']
new_data['SMA100']=SMA100['Adj Close']
new_data
```

```
#Create a function that determines when to buy and sell the stock or the commodity
def execute_order(new_data):
   sigBuy = []
   sigSell = []
   flag=0
   for i in range(len(new_data)):
    if new_data['SMA100'][i]>new_data['SMA50'][i]:
       if flag!=1:
          sigBuy.append(new_data['IBM'][i])
          sigSell.append(np.nan)
          flag=1
       else :
          sigSell.append(np.nan)
          sigBuy.append(np.nan)
    elif new_data['SMA100'][i]<new_data['SMA50'][i]:
       if flag!=-1:
          sigSell.append(new_data['IBM'][i])
          sigBuy.append(np.nan)
          flag=-1
       else :
          sigSell.append(np.nan)
          sigBuy.append(np.nan)
    else:
      sigSell.append(np.nan)
      sigBuy.append(np.nan)

   return (sigBuy,sigSell)
```

```
#Store the buy and sell data into a variable
execute_order=execute_order(new_data)
new_data['Buy']=execute_order[0]
new_data['Sell']=execute_order[1]
```

```
#Visualizing the data with the Strategy
plt.figure(figsize=(12,7))
plt.plot(new_data['IBM'],label='IBM',alpha=0.35)
plt.plot(new_data['SMA50'],label='SMA50',alpha=0.35)
plt.plot(new_data['SMA100'],label='SMA100',alpha=0.35)
plt.scatter(new_data.index,new_data['Buy'],label='Buy',marker='^',color='green')
plt.scatter(new_data.index,new_data['Sell'],label='Sell',marker='v',color='red')
plt.title('IBM Simple Moving Averages Strategy')
plt.xlabel('first of 2015 till the first of 2019')
plt.ylabel('Adj Close Price in $')
plt.legend(loc='upper right')
plt.show()
```
